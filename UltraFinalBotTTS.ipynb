```python
import os
import shutil

folders_to_delete = [
    "D:\\AfterEffectsBot\\Assets\\audios",
    "D:\\AfterEffectsBot\\Assets\\captions",
    "D:\\AfterEffectsBot\\Assets\\descaled images",
    "D:\\AfterEffectsBot\\aebot_AME",
    "D:\\AfterEffectsBot\\Assets\\photos"
]

def delete_folder_contents(folder_path):
    try:
        for root, dirs, files in os.walk(folder_path):
            for file in files:
                file_path = os.path.join(root, file)
                os.remove(file_path)
                print(f"Deleted file: {file_path}")
            for dir in dirs:
                dir_path = os.path.join(root, dir)
                shutil.rmtree(dir_path)
                print(f"Deleted folder: {dir_path}")
    except Exception as e:
        print(f"Error while deleting: {e}")

for folder in folders_to_delete:
    if os.path.exists(folder):
        delete_folder_contents(folder)
    else:
        print(f"Folder not found: {folder}")
```

```python
import os
from scipy.io.wavfile import write as write_wav
from bark.api import semantic_to_waveform
from bark import generate_audio, SAMPLE_RATE
from IPython.display import Audio
import nltk
import numpy as np

nltk.download('punkt')

os.environ["CUDA_VISIBLE_DEVICES"] = "1"

from bark.generation import (generate_text_semantic, preload_models)

preload_models()

# Define the base filename for WAV files along with the desired path
output_directory = "D:\AfterEffectsBot\Assets\\audios"
base_filename = "youtube_audio"

# Initialize a counter for generating unique WAV file names
counter = 1

# Initialize a counter for YouTube text files
youtube_counter = 1

# Initialize a flag to check if there are more YouTube text files
more_youtube_text_files = True

while more_youtube_text_files:
    # Generate a unique WAV file name by appending the counter
    wav_filename = os.path.join(output_directory, f"{base_filename}{counter}.wav")
    
    # Check if the YouTube text file exists
    youtube_text_filename = f"youtube{youtube_counter}.txt"
    if os.path.exists(youtube_text_filename):
        # Read the content from the YouTube text file
        with open(youtube_text_filename, "r", encoding="utf-8") as file:
            youtube_content = file.read().strip()
        
        # Combine YouTube content and generate audio
        script = f"{youtube_content}".replace("\n", " ").strip()

        sentences = nltk.sent_tokenize(script)

        SPEAKER = "v2/en_speaker_6"
        silence = np.zeros(int(0.25 * SAMPLE_RATE))  # quarter second of silence

        pieces = []
        for sentence in sentences:
            audio_array = generate_audio(sentence, history_prompt=SPEAKER)
            pieces += [audio_array, silence.copy()]

        # Concatenate audio pieces
        concatenated_audio = np.concatenate(pieces)

        # Use the unique filename to create the WAV file
        write_wav(wav_filename, SAMPLE_RATE, concatenated_audio)

        # Display the audio
        Audio(concatenated_audio, rate=SAMPLE_RATE)

        # Print the filename to confirm where the WAV file has been saved
        print(f"WAV file {counter} has been saved as {wav_filename}.")

        # Increment counters
        counter += 1
        youtube_counter += 1
    else:
        # If the YouTube text file does not exist, exit the loop
        more_youtube_text_files = False

# Print a message when all YouTube text files have been processed
print("All existing YouTube script files have been processed.")
```
